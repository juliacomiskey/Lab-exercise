```python
from bs4 import BeautifulSoup
import requests
import html5lib
import csv
import pandas as pd


```

```python
source = requests.get("https://www.imdb.com/list/ls055592025/").text
soup = BeautifulSoup(source, "lxml")

new_csv = open ("movies.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(new_csv)
csv_writer.writerow(["Title", "Date of Release", "Length"])

for element in soup.find_all("div", class_="lister-item"):
    title_movie =element.find("h3", class_="lister-item-header").find('a').text.strip()
    date=element.find("span", class_="lister-item-year text-muted unbold").text.strip()
    length= element.find("span", class_="runtime").text.strip()
    
    csv_writer.writerow([title_movie, date, length])
new_csv.close()

sheet2 =pd.read_csv("movies.csv")
sheet2
```

```python
newer_csv = open("cat_scrape.csv", "w",newline="",encoding="utf-8")
csv_writer2 = csv.writer(newer_csv)
csv_writer2.writerow(["Title","Item Description","Hyperlink","Date","Contributor"])

page = 1

while page !=5:
    cats = requests.get(f"https://www.loc.gov/search/?q=cats&sp={page}").text
    soup = BeautifulSoup (cats, "lxml")
                        
    for element in soup.find_all("div", class_="description"):
        titleofcat = element.find("span", class_="item-description-title").text.strip()
        itemdescription = element.find("span", class_="item-description-abstract")
        if itemdescription is not None:
                itemdescription = itemdescription.text.strip()
        else:
                itemdescription = ""
        contributor = element.find("li", class_="contributor")
        if contributor is not None:
            contributor = contributor.find("span").text.strip()
        else:
            contributor = ""
        link = element.find("span", class_="item-description-title").find('a').text.strip()
        date = element.find("li", class_="date")
        if date is not None:
            date = date.find("span").text.strip()
        else:
            date = ""
        
        csv_writer2.writerow([titleofcat, itemdescription,link,date,contributor])
    
    page +=1
                       
sheet4 = pd.read_csv("cat_scrape.csv")
sheet4
```
