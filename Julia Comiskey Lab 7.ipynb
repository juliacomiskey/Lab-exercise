```python
import requests 
import json
import csv
import pandas as pd

lat = "42.098701"
lon = "-75.912537"
genius = requests.get(f"https://api.weather.gov/points/{lat},{lon}")

json_file1 = genius.json()
json_file1["@context"]
forecast = json_file1["properties"]["forecast"]

new_request=requests.get(forecast)

json = new_request.json()

a = json["properties"]["periods"]
a

for i in a:
    name = i["name"]
    temp = i["temperature"]
    detail = i["detailedForecast"]
    print(name)
    print(temp)
    print(detail)
    

```

```python
data_file = open("weather.csv","w", newline="",encoding="utf-8")
csv_writer = csv.writer(data_file)
csv_writer.writerow(["Day","Temp","Description"])

a = json["properties"]["periods"]


for i in a:
    name = i["name"]
    temp = i["temperature"]
    detail = i["detailedForecast"]
    print(name)
    print(temp)
    print(detail)
    csv_writer.writerow([name,temp,detail])
        
data_file.close()
```
